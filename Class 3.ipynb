# Balanced parentheses check

### Given a string of opening and closing parentheses, check whether it is balanced.
### There are 3 types of parentheses here: round brackets (), square brackets [], and curly brackets (braces) {}

__Assumptions__
1. Assume the string doesn't contain any other character other than the ones mentioned above (no spaces, words or numbers).
2. Balanced parentheses require every opening parentheses to be closed in the reverse order opened.

```
Examples:
balance_check("([{}])") should return True
balance_check("{{{}}]") should return False

Similarly:
"()()(){{{}}}" is a balanced string
"()(){]((()))" is not
```

### Instructions
Create a function called "balance_check" that takes a string as its parameter and returns True if it is balanced considering the criteria mentioned above; False otherwise.

````python
#create a stack
#i should first check if the length of the string is odd or even, if its odd, return false
#scan the string, the opening parentheses characters can be pushed to the stack
#once those are pushed, the remaining characters will be closing brackets
#

def balance_check(string):
    stack = []
    
    if int(len(string)) % 2 !== 0:
        return False
    
    for char in string:
        if char in["(", "[", "{"]:
            stack.append(char)
        else:
            top_char == stack.pop()
            if top_char == '(':
                if char != ')':
                    return False
            
````

````python
#professor's output
#each tuple represents a valid match
#line 19: if the stack is now empty here, then that would mean there are no matches and we can return false
#the if statement that has 'in' runs at a big O of N which is a time problem, using a set instead fixes this for both opening and matches
#sets run in constant time

def balancer_check(s):
    if len(s) % 2 != 0:
        return False
    opening = set["([{"]
    matches = set[
        ("(", ")"),
        ("[", "]"),
        ("{", "}")
    ]
    stack = []
    for paren in s:
        if paren in opening:
            stack.append(paren)
        else:
            if stack == []:
                return False
            last_open = stack.pop()
            if (last_open, paren) not in matches:
                return False
    return stack == []
            
````
